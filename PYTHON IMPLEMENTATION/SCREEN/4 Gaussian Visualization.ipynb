```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the means and covariance matrices for 4 Gaussians
means = [
    [1, 1],   # Gaussian 1
    [-1, 1],   # Gaussian 2
    [-1, -1],  # Gaussian 3
    [1, -1]  # Gaussian 4
]

variances = [
    [[3/16, 0], [0, 3/16]],  # Gaussian 1 (Identity covariance)
    [[3/16, 0], [0, 3/16]],  # Gaussian 2 (Correlated variables)
    [[3/16, 0], [0, 3/16]],  # Gaussian 3 (Negative correlation)
    [[3/16, 0], [0, 3/16]]  # Gaussian 4 (Different variances)
]

# Create a grid for plotting
x, y = np.linspace(-2, 2, 100), np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Compute the probability density function for each Gaussian
fig = plt.figure(figsize=(12, 6))

# --- 3D Surface Plot ---
ax1 = fig.add_subplot(121, projection='3d')
Z = np.zeros([100, 100])
for mean, cov in zip(means, variances):
    rv = multivariate_normal(mean, cov)
    Z += rv.pdf(pos)
Z /= 4
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=1)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Probability Density')
ax1.set_title('3D View of 2D Gaussians')

# --- 2D Heatmap ---
ax2 = fig.add_subplot(122)
Z = np.zeros([100, 100])
for mean, cov in zip(means, variances):
    rv = multivariate_normal(mean, cov)
    Z += rv.pdf(pos)
Z/=4
ax2.contourf(X, Y, Z, levels=15, cmap='viridis', alpha=1)

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('2D Contour of 2D Gaussians')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_quadratic(a, b, c):
    x = np.linspace(0, 100, 400)  # Generate 400 points between -10 and 10
    y = a * x**2 + b * x + c  # Quadratic formula
    plt.figure(figsize=(6, 4))
    plt.plot(x, y)
    plt.scatter(50, 2, c="red", label="(Am/2, k*Em)")
    plt.title("Basal Metabolism")
    plt.xlabel("Age")
    plt.ylabel("Energy")
    plt.legend()
    plt.show()

k = 0.02 # PARAMETERS THAT CAN BE TWEAK
Em = 100
Am = 100
a = -(4*k*Em/(Am*Am))
b = +(4*k*Em/(Am))
c = 0
plot_quadratic(a, b, c)  # Plots x^2 - 3x + 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the beta distribution
a = 4  # alpha
b = 8  # beta

# x values between 0 and 1
x = np.linspace(0, 1, 1000)
y = beta.pdf(x, a, b)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', lw=2, label=f'Beta(a={a}, b={b})')
plt.title('Beta Shape')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters
a = 4
b = 8

# Create a grid over [0, 1] x [0, 1]
x = np.linspace(0, 1, 200)
y = np.linspace(0, 1, 200)
X, Y = np.meshgrid(x, y)

# Joint PDF as product of two independent Beta PDFs
Z = beta.pdf(X, a, b) * beta.pdf(Y, a, b)

# Plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_title(f'Bivariate Beta Shape (independent, a={a}, b={b})')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.show()
```
