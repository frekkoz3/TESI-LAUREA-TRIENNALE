```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the means and covariance matrices for 4 Gaussians
means = [
    [1, 1],   # Gaussian 1
    [-1, 1],   # Gaussian 2
    [-1, -1],  # Gaussian 3
    [1, -1]  # Gaussian 4
]

variances = [
    [[3/16, 0], [0, 3/16]],  # Gaussian 1 (Identity covariance)
    [[3/16, 0], [0, 3/16]],  # Gaussian 2 (Correlated variables)
    [[3/16, 0], [0, 3/16]],  # Gaussian 3 (Negative correlation)
    [[3/16, 0], [0, 3/16]]  # Gaussian 4 (Different variances)
]

# Create a grid for plotting
x, y = np.linspace(-2, 2, 100), np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Compute the probability density function for each Gaussian
fig = plt.figure(figsize=(12, 6))

# --- 3D Surface Plot ---
ax1 = fig.add_subplot(121, projection='3d')
Z = np.zeros([100, 100])
for mean, cov in zip(means, variances):
    rv = multivariate_normal(mean, cov)
    Z += rv.pdf(pos)
Z /= 4
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=1)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Probability Density')
ax1.set_title('3D View of 2D Gaussians')

# --- 2D Heatmap ---
ax2 = fig.add_subplot(122)
Z = np.zeros([100, 100])
for mean, cov in zip(means, variances):
    rv = multivariate_normal(mean, cov)
    Z += rv.pdf(pos)
Z/=4
ax2.contourf(X, Y, Z, levels=15, cmap='viridis', alpha=1)

ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('2D Contour of 2D Gaussians')

plt.tight_layout()
plt.show()
```
